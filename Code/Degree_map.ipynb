```python
import requests
import re
from bs4 import BeautifulSoup
```

```python
url = 'https://webapps2.uc.edu/ecurriculum/DegreePrograms/Home/MajorMap/11017'
response = requests.get(url)
if response:
    print(response.status_code)

soup = BeautifulSoup(response.text, 'html.parser')

col_val = 2
rows=soup.select('div.col-md-4.form-group.mm-col-{}'.format(col_val))

Cred_req = False
Cred_req_regex = '(Minimum Credits: </a>)(\d* - \d* )(Credits)'

Duration = False
Duration_regex = '(<span>Duration: )(\d*)( Years</span><br/>)'

sem_regex = '(<h2>)(Fall \d{4}|Spring \d{4}|Summer \d{4})(</h2>)'
class_regex = '<td>(\w*\d{4}\w{0,1})</td>'
gen_regex = 'General Education'
CS_elective = ''
direct_elective = ''
science_lab = ''


CS_Degree_map = {}
```

```python
if not(Cred_req):
        hits = re.findall(Cred_req_regex, str(rows))
        if hits:
            Cred_req = hits[0][1]
if not(Duration):
    hits = re.findall(Duration_regex, str(rows))
    if hits:
        Duration = hits[0][1]

while rows!=[]:
    for row in rows:
        sem_perhaps = re.findall(sem_regex, str(row))
        if sem_perhaps:
            CS_Degree_map[sem_perhaps[0][1]] = []
            key = sem_perhaps[0][1]
        
        class_perhaps = re.findall(class_regex, str(row))
        gen_hits = re.findall(gen_regex, str(row))

        if class_perhaps:
             CS_Degree_map[key] = class_perhaps + gen_hits
    col_val+=1
    rows=soup.select('div.col-md-4.form-group.mm-col-{}'.format(col_val))

```
