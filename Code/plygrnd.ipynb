```python
from bs4 import BeautifulSoup

# Read the HTML (the .txt file) as a string
with open("CS_Fall2025.txt", "r", encoding="utf-8") as file:
    html_content = file.read()

soup = BeautifulSoup(html_content, 'html.parser')

class_hash_map = {}

# Depending on the file's structure, identify the table or divs containing class info
rows = soup.select('table.PSLEVEL1GRIDNBONBO tr')  # or an appropriate selector
# rows = soup.select('td.PSGROUPBOXLABEL')


for row in rows[1:]:  # skip header
    columns = row.find_all('td')
    if not columns or len(columns) < 12:
        continue
    class_id = columns[0].get_text(strip=True)
    section = columns[1].get_text(strip=True)
    credits = columns[2].get_text(strip=True)
    attributes = columns[3].get_text(strip=True) if columns[3].get_text(strip=True) else None
    bok_bacc = columns[4].get_text(strip=True)
    campus = columns[5].get_text(strip=True)
    days_times = columns[6].get_text(strip=True)
    instruction_method = columns[7].get_text(strip=True)
    room = columns[8].get_text(strip=True)
    offering_college = columns[9].get_text(strip=True)
    instructor = columns[10].get_text(strip=True)
    meeting_dates = columns[11].get_text(strip=True)

    class_hash_map[(int(class_id))] = {
        'Section': section,
        'Credits': credits,
        'Attributes': attributes,
        'BOK_BACC': bok_bacc,
        'Campus': campus,
        'Days_Times': days_times,
        'Instruction_Method': instruction_method,
        'Room': room,
        'Offering_College': offering_college,
        'Instructor': instructor,
        'Meeting_Dates': meeting_dates,
    }

class_hash_map
```

```python
import re
from bs4 import BeautifulSoup

CS_data = {}

with open("CS_Fall2025.txt", "r", encoding="utf-8") as file:
    html_content = file.read()

soup = BeautifulSoup(html_content, 'html.parser')

class_title_count = 0
title_regex = '(CS \d{4}.*)( - .*)'
class_nbr_regex = '\d{5}'
links = soup.find_all("a")
for link in links:
    class_title_perhaps = link.get_attribute_list('title')[0]
    
    if class_title_perhaps != None:
        match = re.findall(title_regex, str(class_title_perhaps))
        try:
            class_id = match[0][0]
            title = match[0][0] + match[0][1]
            CS_data[class_id] = {'Name' : title}
        except:
            continue
    
    class_number_perhaps = link.getText()
    
    if class_number_perhaps!=None:
        match = re.findall(class_nbr_regex, str(class_number_perhaps))
        if match != []:
            # print(class_number_perhaps, match)
            if class_id:
                CS_data[class_id][int(match[0])] = {}

len(CS_data)
```

```python
for CS_class in CS_data:
    classes = list(CS_data[CS_class].keys())[1:]
    for class_ in classes:
        CS_class_attributes = class_hash_map[class_]
        CS_data[CS_class][class_] = CS_class_attributes

CS_data
```
